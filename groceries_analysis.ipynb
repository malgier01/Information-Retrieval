You shall already be familiar with the concept of association rules and the apriori algorithm. Association rule mining is a method for discovering patterns within large data sets. It focuses on identifying relationships between variables and leveraging those connections to make predictions or informed decisions. The primary objective is to uncover rules that reveal the associations between various items in the data.



### Task
Load data from data.txt file - it contains lists of grocery shopping done by nearly 2000 customers.
Store it in a boolean one hot encoded dataframe - True for items bought in a given transaction, False otherwise.

```python
import re
groceries = {}
for line in open("data.txt"):
  line = re.sub(r'[^a-zA-Z\s]', '', line)
  line = line.split()
  for item in line:
    if item in groceries:
      groceries[item] += 1
    else:
      groceries[item] = 1
groceries
```

```python
groceries_df = pd.DataFrame(columns=groceries.keys())
groceries_df
for line in open("data.txt"):
  line = re.sub(r'[^a-zA-Z\s]', '', line)
  vector = []
  for key in groceries:
    if key in line:
      vector.append(True)
    else:
      vector.append(False)
  vector_df = pd.DataFrame([vector], columns=groceries.keys())
  groceries_df = pd.concat([groceries_df, vector_df], ignore_index=True)
groceries_df
```

To extract rules you can use e.g. apriori algorithm implemented in mlxtend. There are other algorithms performing the same task but using different approaches e.g. fpgrowth internally uses a tree-based structure which makes it faster in most real-life examples.

### Task

```python
from mlxtend.frequent_patterns import apriori, association_rules

items = apriori(groceries_df, min_support=0.05, use_colnames=True)
rules = association_rules(items, metric="confidence", min_threshold=0.7, num_itemsets = len(items))
```

```python
# Disjunction
import itertools

def disjunctive_support(itemset, transactions):
    count = transactions[list(itemset)].sum(axis=1).astype(bool).sum()
    return count / len(transactions)


def generate_disjunctive_itemsets(items, k):
    return list(itertools.combinations(items, k))


def apriori_disjunctive(transactions, min_support):
    items = transactions.columns
    frequent_itemsets = []
    k = 2
    disjunctive_itemsets = generate_disjunctive_itemsets(items, k)
    for itemset in disjunctive_itemsets:
        support = disjunctive_support(itemset, transactions)
        if support >= min_support:
            frequent_itemsets.append((itemset, support))

    return frequent_itemsets

min_support = 0.6
frequent_disjunctive_itemsets = apriori_disjunctive(groceries_df, min_support)
frequent_disjunctive_itemsets
```

```python
#expected co-occurrence vs recorded co-occurr

import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

item_probabilities = groceries_df.mean()
joint_probabilities = groceries_df.T @ groceries_df / len(groceries_df)

def expected_co_occurrence(item1, item2, item_probabilities):
    return item_probabilities[item1] * item_probabilities[item2]

def chi_square_test(item1, item2, groceries_df):
    contingency_table = pd.crosstab(groceries_df[item1], groceries_df[item2])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    return chi2, p

def calculate_lift(item1, item2, joint_probabilities, item_probabilities):
    return joint_probabilities.loc[item1, item2] / (item_probabilities[item1] * item_probabilities[item2])


significant_rules = []
for item1 in groceries_df.columns:
    for item2 in groceries_df.columns:
        if item1 != item2:
            observed = joint_probabilities.loc[item1, item2]
            expected = expected_co_occurrence(item1, item2, item_probabilities)

            chi2, p_value = chi_square_test(item1, item2, groceries_df)

            lift = calculate_lift(item1, item2, joint_probabilities, item_probabilities)

            if p_value < 0.05 and lift > 1:
                significant_rules.append({
                    'Item1': item1,
                    'Item2': item2,
                    'Observed': observed,
                    'Expected': expected,
                    'Lift': lift,
                    'P-value': p_value
                })

significant_rules_groceries_df = pd.DataFrame(significant_rules)
print(significant_rules_groceries_df)
```
